#### 3. Faça um Programa que peça dois números e imprima a soma.

```python
numero1 = int(input('Digite um numero:'))
numero2 = int(input('Digite outro numero:'))
soma = numero1 + numero2
print(f'A soma de {numero1} + {numero2} é soma{soma}')
```

#### 4. Faça um Programa que peça as 4 notas bimestrais de um aluno e mostre a média de todas as notas.

```python
nota1 = float(input('Digite a nota do primeiro bimestre'))
nota2 = float(input('Digite a nota do segundo bimestre'))
nota3 = float(input('Digite a nota do terceiro bimestre'))
nota4 = float(input('Digite a nota do quarto bimestre'))
media = (nota1 + nota2 + nota3 + nota4) / 4
print(f'Sua notas {nota1} , {nota2} , {nota3} , {nota4} ,Sua media foi: {media}')
```

#### 5. Faça um Programa que converta metros para centímetros. Você pode pedir o comprimento em metros para o usuário (input).

```python
metros = float(input('Digite uma altura em metros: '))
centimetros = metros * 100

print(f'Voce digitou a altura {metros}M, convertido em centimetros :{centimetros}CM')
```

#### 6. Faça um Programa que calcule a área de uma sala de um apartamento. Para isso, o seu programa precisa pedir a largura da sala, o comprimento da sala e imprimir a área em m² da sala.

```python
largura_sala = float(input('Dgite a largura da sala:'))
comprimento_sala = float(input('Dgite a comprimento da sala:'))
area_sala = largura_sala * comprimento_sala

print(f'A largura da sala :{largura_sala}M o comprimento da sala{comprimento_sala}M sua area total:{area_sala}M² ')
```

#### 7. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.

```python
preco_hora = float(input('Qual o preco da sua hora trabalhada:'))
qtde_hora = float(input('Quantas  horas trabalhada:'))
salario_mes = preco_hora * qtde_hora

print(f'Preco da hora {preco_hora}R$ quantidade horas: {qtde_hora}H salario do mes: {salario_mes}R$')
```

#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
temp_fahr = float(input('Digite a temperatura em Fahrenheit:'))
converte_temp_celsius = 5/9*(temp_fahr - 32)

print(f'A temperatura fahrnheit {temp_fahr} convertida temperatura Celsius: {converte_temp_celsius}')
```
